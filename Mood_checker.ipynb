```python
mood = input("How is your mood today?(good/sad)")

if mood.lower() == "good":
  print("Awesome! keep learning and stay focused.")
elif mood.lower() == "sad":
  print("It's okay, this too shall pass. Take 10-minute walk and then get back to python.")
else:
  print("Just remember your goal and keep moving forward!")
```
